Visualize Distributions of Numerical Features:

Visualize Relationships between Features and Target:

Scatter Plots (Numerical vs. Target):

```python
print("Scatter plots: Numerical vs Profit")
L=['R&D Spend', 'Administration', 'Marketing Spend']
for col in L:
    plt.figure(figsize=(8,5))
    sns.scatterplot(x=df[col], y=df['Profit'])
    plt.title(f'{col} vs. Profit')
    plt.xlabel(col)
    plt.ylabel('Profit')
    plt.show()
```

Conclusion: R&D Spend have linear relationship with profit followed by Marketing Spend

### Step 3: Split the data into Feature(X) and Target(Y)

```python
x=df[['R&D Spend', 'Administration', 'Marketing Spend']]
y=df.Profit
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
x_train
```

```python
x_train.shape
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(x_train,y_train)
```

```python
y_predict=LR.predict(x_test)
y_predict  #Predict Value
```

```python
y_test   #Actual Value
```

### Model Evaluation

```python
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error, r2_score
```

```python
mse_value=mean_squared_error(y_test,y_predict)
print(f'mean_squared_error, {mse_value:.2f}')
```

```python
rmse_value=root_mean_squared_error(y_test,y_predict)
print(f'root_mean_squared_error, {rmse_value:.2f}')
```

```python
r2_value = r2_score(y_test,y_predict)
print(f'r2_score, {r2_value:.2f}')
```

```python
mae_value=mean_absolute_error(y_test,y_predict)
print(f'mean_absolute_error, {mae_value:.2f}')
```
